```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Function Input
# ==========================
print("=== Trapezoidal Rule Integration ===")
func_str = input("Enter the function f(x) (e.g., 1/(1+x**2)): ") or "1/(1+x**2)"
try:
    f = lambda x: eval(func_str)
except Exception as e:
    print("❌ Invalid function!")
    exit()

try:
    a = float(input("Enter lower limit a (default 0): ") or "0")
    b = float(input("Enter upper limit b (default 6): ") or "6")
    n = int(input("Enter number of subintervals n (default 1): ") or "1")
except ValueError:
    print("❌ Invalid input! Please enter numeric values.")
    exit()

# ==========================
# Trapezoidal Rule
# ==========================
def trapezoidal_rule(f, a, b, n):
    """Composite Trapezoidal Rule"""
    x = np.linspace(a, b, n+1)
    h = (b - a) / n
    I = h * (0.5*f(a) + 0.5*f(b) + sum(f(x[1:-1])))
    return I, x

I, x_trap = trapezoidal_rule(f, a, b, n)

print(f"\nTrapezoidal Rule Approximation: {I:.6f}")

# ==========================
# Plot
# ==========================
x_plot = np.linspace(a, b, 200)
y_plot = f(x_plot)

plt.figure(figsize=(8,5))
plt.plot(x_plot, y_plot, 'b', label='f(x)')
for i in range(n):
    xs = [x_trap[i], x_trap[i], x_trap[i+1], x_trap[i+1]]
    ys = [0, f(x_trap[i]), f(x_trap[i+1]), 0]
    plt.fill(xs, ys, color='skyblue', alpha=0.4)
plt.title(f"Trapezoidal Rule Approximation with n={n}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Function Input
# ==========================
print("=== Simpson's 1/3 Rule Integration ===")
func_str = input("Enter the function f(x) (e.g., 1/(1+x**2)): ") or "1/(1+x**2)"
try:
    f = lambda x: eval(func_str)
except Exception as e:
    print("❌ Invalid function!")
    exit()

try:
    a = float(input("Enter lower limit a (default 0): ") or "0")
    b = float(input("Enter upper limit b (default 6): ") or "6")
    n = int(input("Enter number of subintervals n (even, default 2): ") or "2")
    if n % 2 != 0:
        print("⚠️ n must be even. Incrementing n by 1.")
        n += 1
except ValueError:
    print("❌ Invalid input! Please enter numeric values.")
    exit()

# ==========================
# Simpson's 1/3 Rule (Composite)
# ==========================
def simpson_13_rule(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    I = y[0] + y[-1] + 4*sum(y[1:n:2]) + 2*sum(y[2:n-1:2])
    I *= h/3
    return I, x, y

I, x_sim, y_sim = simpson_13_rule(f, a, b, n)

print(f"\nSimpson's 1/3 Rule Approximation: {I:.6f}")

# ==========================
# Plotting
# ==========================
x_plot = np.linspace(a, b, 200)
y_plot = f(x_plot)

plt.figure(figsize=(8,5))
plt.plot(x_plot, y_plot, 'g', label='f(x)')
plt.fill_between(x_plot, y_plot, color='lightgreen', alpha=0.4)
plt.scatter(x_sim, y_sim, color='red')  # plot nodes
plt.title(f"Simpson's 1/3 Rule Approximation with n={n}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Function Input
# ==========================
print("=== Simpson's 3/8 Rule Integration ===")
func_str = input("Enter the function f(x) (e.g., 1/(1+x**2)): ") or "1/(1+x**2)"
try:
    f = lambda x: eval(func_str)
except Exception as e:
    print("❌ Invalid function!")
    exit()

try:
    a = float(input("Enter lower limit a (default 0): ") or "0")
    b = float(input("Enter upper limit b (default 6): ") or "6")
    n = int(input("Enter number of subintervals n (multiple of 3, default 3): ") or "3")
    if n % 3 != 0:
        print("⚠️ n must be a multiple of 3. Adjusting n to the next multiple.")
        n += 3 - n % 3
except ValueError:
    print("❌ Invalid input! Please enter numeric values.")
    exit()

# ==========================
# Simpson's 3/8 Rule (Composite)
# ==========================
def simpson_38_rule(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    I = y[0] + y[-1] + 3*sum(y[1:n:3] + y[2:n:3]) + 2*sum(y[3:n-2:3])
    I *= 3*h/8
    return I, x, y

I, x_sim, y_sim = simpson_38_rule(f, a, b, n)

print(f"\nSimpson's 3/8 Rule Approximation: {I:.6f}")

# ==========================
# Plotting
# ==========================
x_plot = np.linspace(a, b, 200)
y_plot = f(x_plot)

plt.figure(figsize=(8,5))
plt.plot(x_plot, y_plot, 'r', label='f(x)')
plt.fill_between(x_plot, y_plot, color='salmon', alpha=0.4)
plt.scatter(x_sim, y_sim, color='red')  # plot nodes
plt.title(f"Simpson's 3/8 Rule Approximation with n={n}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
