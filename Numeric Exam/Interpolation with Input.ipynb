```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# ==========================
# User Input Section
# ==========================
print("=== Newton's Forward Interpolation ===")
try:
    x_input = input("Enter x values separated by commas (e.g., 1,2,3,4,5): ")
    y_input = input("Enter corresponding y values separated by commas (e.g., 1,8,27,64,125): ")
    X = float(input("Enter the point X to interpolate (Ex:2.5): "))
    
    x = np.array([float(i) for i in x_input.strip().split(",")])
    y = np.array([float(i) for i in y_input.strip().split(",")])
    
    if len(x) != len(y):
        print("❌ x and y must have the same number of elements.")
        exit()
except Exception as e:
    print("❌ Invalid input!")
    exit()

# ==========================
# Forward Difference Table
# ==========================
n = len(x)
diff = np.zeros((n,n))
diff[:,0] = y

for i in range(1,n):
    for j in range(n-i):
        diff[j,i] = diff[j+1,i-1] - diff[j,i-1]

# ==========================
# Newton Forward Interpolation
# ==========================
h = x[1] - x[0]
u = (X - x[0]) / h

I = y[0]
for i in range(1,n):
    term = u
    for j in range(1,i):
        term *= (u-j)
    I += term * diff[0,i] / factorial(i)

print(f"\nInterpolated value at X={X}: {I:.6f}")

# ==========================
# Plotting
# ==========================
plt.figure(figsize=(8,5))
plt.plot(x, y, 'bo-', label='Data points')
plt.scatter(X, I, color='red', s=80, label=f'Interpolated at X={X}')
plt.title("Newton's Forward Interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# ==========================
# User Input Section
# ==========================
print("=== Newton's Backward Interpolation ===")
try:
    x_input = input("Enter x values separated by commas (e.g., 1,2,3,4,5): ")
    y_input = input("Enter corresponding y values separated by commas (e.g., 1,8,27,64,125): ")
    X = float(input("Enter the point X to interpolate(Ex:4.5): "))
    
    x = np.array([float(i) for i in x_input.strip().split(",")])
    y = np.array([float(i) for i in y_input.strip().split(",")])
    
    if len(x) != len(y):
        print("❌ x and y must have the same number of elements.")
        exit()
except Exception as e:
    print("❌ Invalid input!")
    exit()

# ==========================
# Backward Difference Table
# ==========================
n = len(x)
diff = np.zeros((n,n))
diff[:,0] = y

for i in range(1,n):
    for j in range(i,n):
        diff[j,i] = diff[j,i-1] - diff[j-1,i-1]

# ==========================
# Newton Backward Interpolation
# ==========================
h = x[1] - x[0]
u = (X - x[-1]) / h

I = y[-1]
for i in range(1,n):
    term = u
    for j in range(1,i):
        term *= (u+j)
    I += term * diff[-1,i] / factorial(i)

print(f"\nInterpolated value at X={X}: {I:.6f}")

# ==========================
# Plotting
# ==========================
plt.figure(figsize=(8,5))
plt.plot(x, y, 'go-', label='Data points')
plt.scatter(X, I, color='red', s=80, label=f'Interpolated at X={X}')
plt.title("Newton's Backward Interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Lagrange Interpolation Function
# ==========================
def lagrange(x_data, y_data, X):
    n = len(x_data)
    result = 0
    for i in range(n):
        term = y_data[i]
        for j in range(n):
            if j != i:
                term *= (X - x_data[j])/(x_data[i]-x_data[j])
        result += term
    return result

# ==========================
# User Input Section
# ==========================
print("=== Lagrange Interpolation ===")
try:
    x_input = input("Enter x values separated by commas (e.g., 1,2,3,4,5): ")
    y_input = input("Enter corresponding y values separated by commas (e.g., 1,8,27,64,125): ")
    X = float(input("Enter the point X to interpolate(Ex:2.5): "))
    
    x = np.array([float(i) for i in x_input.strip().split(",")])
    y = np.array([float(i) for i in y_input.strip().split(",")])
    
    if len(x) != len(y):
        print("❌ x and y must have the same number of elements.")
        exit()
except Exception as e:
    print("❌ Invalid input!")
    exit()

# ==========================
# Compute Interpolation
# ==========================
I = lagrange(x, y, X)
print(f"\nLagrange Interpolation at X={X}: {I:.6f}")

# ==========================
# Plotting
# ==========================
plt.figure(figsize=(8,5))
plt.plot(x, y, 'm', marker='o', label='Data points')
plt.scatter(X, I, color='red', s=80, label=f'Interpolated at X={X}')
plt.title("Lagrange Interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
