```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- Take input from user ---
n = int(input("How many data points do you want to enter? "))

X = []
y = []

print("\nEnter values for X and Y:")
for i in range(n):
    x_val = float(input(f"X[{i+1}] = "))
    y_val = float(input(f"Y[{i+1}] = "))
    X.append(x_val)
    y.append(y_val)

# Convert to NumPy arrays and reshape X
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# --- Train linear regression model ---
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# --- Display results ---
print("\nEquation of line:  y = {:.3f}x + {:.3f}".format(model.coef_[0], model.intercept_))
print("Predicted values:", y_pred)

# --- Plot ---
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title("Linear Regression (User Input Data)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# --- Take user input ---
n = int(input("How many data points do you want to enter? "))

X = []
y = []

print("\nEnter values for X and Y:")
for i in range(n):
    x_val = float(input(f"X[{i+1}] = "))
    y_val = float(input(f"Y[{i+1}] = "))
    X.append(x_val)
    y.append(y_val)

degree = int(input("\nEnter the degree of the polynomial: "))

# --- Convert to numpy arrays ---
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# --- Polynomial Regression ---
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

model = LinearRegression().fit(X_poly, y)
y_pred = model.predict(X_poly)

# --- Display results ---
coefs = model.coef_
intercept = model.intercept_
print("\nPolynomial Coefficients:")
for i, c in enumerate(coefs):
    print(f"  Coefficient of x^{i}: {c:.3f}")
print(f"Intercept: {intercept:.3f}")

# --- Sort X for smoother plot ---
X_sorted, y_pred_sorted = zip(*sorted(zip(X.flatten(), y_pred)))

# --- Plot ---
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_sorted, y_pred_sorted, color='red', label='Polynomial Fit')
plt.title(f"Polynomial Regression (degree = {degree})")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- Take input from user ---
n = int(input("How many data points do you want to enter? "))
m = int(input("How many features (independent variables) per data point? "))

# --- Enter X values ---
X = []
y = []

print("\nEnter the data points:")
for i in range(n):
    row = []
    print(f"\nData point {i+1}:")
    for j in range(m):
        val = float(input(f"  Feature {j+1}: "))
        row.append(val)
    X.append(row)
    y_val = float(input("  Target (Y): "))
    y.append(y_val)

# --- Convert to NumPy arrays ---
X = np.array(X)
y = np.array(y)

# --- Train the model ---
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# --- Display results ---
print("\nCoefficients (weights for each feature):")
for i, c in enumerate(model.coef_):
    print(f"  Feature {i+1}: {c:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

# --- Plot results ---
plt.scatter(range(len(y)), y, label='Actual', color='blue')
plt.plot(range(len(y_pred)), y_pred, color='red', label='Predicted')
plt.title("Multiple Linear Regression (User Input Data)")
plt.xlabel("Data Point Index")
plt.ylabel("Y Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# --- Take user input ---
n = int(input("How many data points do you want to enter? "))

X = []
y = []

print("\nEnter values for X and Y (Y must be 0 or 1):")
for i in range(n):
    x_val = float(input(f"X[{i+1}] = "))
    y_val = int(input(f"Y[{i+1}] = "))
    X.append(x_val)
    y.append(y_val)

# Convert to NumPy arrays and reshape X
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# --- Train logistic regression model ---
model = LogisticRegression().fit(X, y)
y_pred = model.predict(X)
y_prob = model.predict_proba(X)[:, 1]

# --- Display results ---
print("\nPredicted labels:", y_pred)
print("Predicted probabilities:", np.round(y_prob, 3))
print(f"Model Coefficient: {model.coef_[0][0]:.3f}")
print(f"Model Intercept: {model.intercept_[0]:.3f}")

# --- Plot ---
plt.figure(figsize=(7,4))
plt.scatter(X, y, color='blue', label='Actual (Y)')
plt.scatter(X, y_pred, color='red', marker='x', label='Predicted (Y_pred)')
plt.plot(X, y_prob, color='green', label='Probability curve')
plt.title("Logistic Regression (User Input Data)")
plt.xlabel("X values")
plt.ylabel("Y / Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor

# --- Take user input ---
n = int(input("How many data points do you want to enter? "))

X = []
y = []

print("\nEnter values for X and Y:")
for i in range(n):
    x_val = float(input(f"X[{i+1}] = "))
    y_val = float(input(f"Y[{i+1}] = "))
    X.append(x_val)
    y.append(y_val)

# --- User configuration for SGD parameters ---
max_iter = int(input("\nEnter maximum iterations (e.g., 1000): "))
tol = float(input("Enter tolerance (e.g., 1e-3): "))
eta0 = float(input("Enter learning rate (eta0, e.g., 0.01): "))

# --- Convert to NumPy arrays ---
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# --- Train the model ---
model = SGDRegressor(max_iter=max_iter, tol=tol, eta0=eta0, learning_rate='constant', random_state=42)
model.fit(X, y)
y_pred = model.predict(X)

# --- Display results ---
print("\nSGD Regression Results:")
print(f"  Coefficient (slope): {model.coef_[0]:.4f}")
print(f"  Intercept: {model.intercept_[0]:.4f}")

# --- Plot results ---
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.title("SGD Regression (User Input Data)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```
